# Full Transformer Toy Pipeline: Q/K/V → MHA → Block → Logits → Decoding + Heatmaps

1. Multi-Head Self-Attention (Q/K/V) with **causal masking**
2. **Residuals**, **LayerNorm**, and **Feed-Forward (FFN, GELU)** (pre-LN block)
3. Output head to **logits** over a toy vocabulary → **probabilities**
4. **Decoding demos**: Greedy (deterministic), Temperature, Top-k, Top-p (stochastic)
5. **Attention heatmaps** (matplotlib only, one chart per cell)

Everything is tiny and deterministic given the RNG seed so you can inspect shapes and values.

```python
import numpy as np
import math
import matplotlib.pyplot as plt
np.set_printoptions(precision=3, suppress=True)

def softmax(x, axis=-1):
    x = x - np.max(x, axis=axis, keepdims=True)
    e = np.exp(x)
    return e / np.sum(e, axis=axis, keepdims=True)

def gelu(x):
    return 0.5 * x * (1.0 + np.tanh( (2.0/np.pi)**0.5 * (x + 0.044715 * (x**3)) ))

class LayerNorm:
    def __init__(self, d_model, eps=1e-5):
        self.gamma = np.ones((d_model,), dtype=float)
        self.beta  = np.zeros((d_model,), dtype=float)
        self.eps   = eps
    def __call__(self, x):
        mean = x.mean(axis=-1, keepdims=True)
        var  = ((x - mean)**2).mean(axis=-1, keepdims=True)
        xhat = (x - mean) / np.sqrt(var + self.eps)
        return self.gamma * xhat + self.beta

rng = np.random.default_rng(8675309)  # deterministic seed
```

## 1) Tiny configuration and embeddings
We create token and positional embeddings and build the input tensor `X`.

```python
seq_len   = 5
vocab_sz  = 12
d_model   = 8
num_heads = 2
d_head    = d_model // num_heads  # 4 per head

# toy token ids (pretend from tokenizer)
tokens = np.array([2, 5, 7, 3, 9])
print('Tokens:', tokens)

token_embed_table = rng.normal(0, 1, size=(vocab_sz, d_model))
pos_embed_table   = rng.normal(0, 0.1, size=(seq_len, d_model))

print('token_embed_table', token_embed_table)

X = token_embed_table[tokens] + pos_embed_table  # (seq_len, d_model)
print('X shape:', X.shape)
print(X)
```

## 2) Multi-Head Self-Attention with causal mask
We implement per-head Q/K/V projections, softmax attention with a causal mask, then concatenate heads and project with `W_O`.

```python
# Per-head Q/K/V and output projection
W_Q = rng.normal(0, 0.5, size=(num_heads, d_model, d_head))
W_K = rng.normal(0, 0.5, size=(num_heads, d_model, d_head))
W_V = rng.normal(0, 0.5, size=(num_heads, d_model, d_head))
b_Q = rng.normal(0, 0.05, size=(num_heads, d_head))
b_K = rng.normal(0, 0.05, size=(num_heads, d_head))
b_V = rng.normal(0, 0.05, size=(num_heads, d_head))

W_O = rng.normal(0, 0.5, size=(num_heads*d_head, d_model))
b_O = rng.normal(0, 0.05, size=(d_model,))

def mha_with_weights(x):
    # x: (seq_len, d_model)
    Q = np.zeros((num_heads, seq_len, d_head))
    K = np.zeros((num_heads, seq_len, d_head))
    V = np.zeros((num_heads, seq_len, d_head))
    for h in range(num_heads):
        Q[h] = x @ W_Q[h] + b_Q[h]
        K[h] = x @ W_K[h] + b_K[h]
        V[h] = x @ W_V[h] + b_V[h]
    scale = math.sqrt(d_head)
    attn_weights = []
    contexts = []
    for h in range(num_heads):
        scores = (Q[h] @ K[h].T) / scale
        mask = np.triu(np.ones_like(scores), k=1)
        scores = np.where(mask==1, -1e9, scores)
        attn = softmax(scores, axis=-1)
        context = attn @ V[h]
        attn_weights.append(attn)
        contexts.append(context)
    concat = np.concatenate(contexts, axis=-1)  # (seq_len, d_model)
    out = concat @ W_O + b_O
    return out, attn_weights  # attn_weights: list of (seq_len, seq_len) per head
```

## 3) One pre-LN Transformer block: Residual + LN + MHA, then Residual + LN + FFN

```python
d_ff = 16
W1 = rng.normal(0, 0.5, size=(d_model, d_ff))
b1 = rng.normal(0, 0.05, size=(d_ff,))
W2 = rng.normal(0, 0.5, size=(d_ff, d_model))
b2 = rng.normal(0, 0.05, size=(d_model,))

ln1 = LayerNorm(d_model)
ln2 = LayerNorm(d_model)

def transformer_block(x):
    # MHA sublayer
    y_norm = ln1(x)
    y, attn_weights = mha_with_weights(y_norm)
    x = x + y  # residual
    # FFN sublayer
    z_norm = ln2(x)
    z = gelu(z_norm @ W1 + b1)
    z = z @ W2 + b2
    x = x + z  # residual
    return x, attn_weights
```

## 4) Run the block to get contextual hidden states

```python
H, attn_weights = transformer_block(X)
print('H shape:', H.shape)
print(H)
print('Heads:', len(attn_weights), '| Each shape:', attn_weights[0].shape)
```

## 5) Attention heatmaps
Heatmaps for Head 0 and Head 1. Rows = query positions; columns = key positions.

```python
plt.figure()
plt.imshow(attn_weights[0], aspect='auto')
plt.title('Head 0 — Attention Weights')
plt.xlabel('Key index')
plt.ylabel('Query index')
plt.colorbar()
plt.show()
```

```python
plt.figure()
plt.imshow(attn_weights[1], aspect='auto')
plt.title('Head 1 — Attention Weights')
plt.xlabel('Key index')
plt.ylabel('Query index')
plt.colorbar()
plt.show()
```

## 6) Output head → logits over vocabulary
We map the **last position** hidden state to logits over the toy vocabulary and get probabilities.

```python
W_vocab = rng.normal(0, 0.4, size=(d_model, vocab_sz))
b_vocab = rng.normal(0, 0.05, size=(vocab_sz,))

h_last = H[-1]  # (d_model,)
logits = h_last @ W_vocab + b_vocab  # (vocab_sz,)
probs  = softmax(logits)

print('logits shape:', logits.shape)
print('probs shape:', probs.shape, '| sum:', probs.sum())
print('Top-5 token indices and probs:')
for i in np.argsort(-probs)[:5]:
    print(int(i), float(probs[i]))
```

## 7) Decoding: Greedy vs Temperature / Top-k / Top-p

```python
def greedy_decode(probs):
    return int(np.argmax(probs))

def sample_with_temperature(logits, temperature=1.0, rng=None):
    if rng is None:
        rng = np.random.default_rng()
    scaled = logits / max(temperature, 1e-8)
    p = softmax(scaled)
    return int(rng.choice(len(p), p=p))

def top_k_logits(logits, k):
    if k <= 0 or k >= len(logits):
        return logits.copy()
    idx = np.argsort(logits)
    cutoff = logits[idx[-k]]
    out = np.where(logits < cutoff, -1e9, logits)
    return out

def top_p_logits(logits, p=0.9):
    probs_ = softmax(logits)
    idx_sorted = np.argsort(-probs_)
    cum = 0.0
    keep = []
    for i in idx_sorted:
        keep.append(i)
        cum += probs_[i]
        if cum >= p:
            break
    mask = np.ones_like(logits, dtype=bool)
    mask[keep] = False
    out = logits.copy()
    out[mask] = -1e9
    return out

rng_dec = np.random.default_rng(20251001)

greedy_id = greedy_decode(probs)
temp10_id = sample_with_temperature(logits, temperature=1.0, rng=rng_dec)
temp07_id = sample_with_temperature(logits, temperature=0.7, rng=rng_dec)
temp13_id = sample_with_temperature(logits, temperature=1.3, rng=rng_dec)

k3_logits = top_k_logits(logits, k=3)
k3_id = sample_with_temperature(k3_logits, temperature=1.0, rng=rng_dec)

p90_logits = top_p_logits(logits, p=0.9)
p90_id = sample_with_temperature(p90_logits, temperature=1.0, rng=rng_dec)

print('Greedy (deterministic):', greedy_id)
print('Sampling (temp=1.0):  ', temp10_id)
print('Sampling (temp=0.7):  ', temp07_id)
print('Sampling (temp=1.3):  ', temp13_id)
print('Top-k=3 sampling:     ', k3_id)
print('Top-p=0.90 sampling:  ', p90_id)
```

## 8) Multiple samples to visualize variability

```python
rng_multi = np.random.default_rng(4242)

def samples(method, n=5):
    out = []
    for _ in range(n):
        if method=='temp1.0':
            out.append(sample_with_temperature(logits, 1.0, rng_multi))
        elif method=='temp0.0':
            out.append(sample_with_temperature(logits, 0.0, rng_multi))
        elif method=='topk3':
            out.append(sample_with_temperature(top_k_logits(logits,3), 1.0, rng_multi))
        elif method=='topp0.9':
            out.append(sample_with_temperature(top_p_logits(logits,0.9), 1.0, rng_multi))
    return out

print('Greedy (always same):', greedy_id)
print('Temp=0.0 samples:   ', samples('temp0.0')) ## Can't expect much variation on a small sample here. :-(
print('Temp=1.0 samples:   ', samples('temp1.0'))
print('Top-k=3 samples:    ', samples('topk3'))
print('Top-p=0.9 samples:  ', samples('topp0.9'))
```
